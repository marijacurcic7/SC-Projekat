```python
def extract_features(file_name):
    try:
        audio, sample_rate = librosa.load(file_name, res_type="kaiser_fast")
        mfccs = librosa.feature.mfcc(y=audio, sr=22050, n_mfcc=40)
        mfccscaled = np.mean(mfccs.T, axis=0)
    
    except Exception as e:
        print("Error encountered while parsing file: ", file_name)
        return None
    
    return mfccscaled
```

```python
folder_path = "dataset/Accordion/"
features = []

from os import listdir
from os.path import isfile, join

onlyfiles = [f for f in listdir(folder_path) if isfile(join(folder_path, f))]

for file in onlyfiles: 
    file_name = folder_path + file
    a = file.split("-")
    instrument = a[0]
    pitch = a[2]
    if(len(pitch) == 3):
        pitch = pitch[:2]
    else:
        pitch = pitch[:1]
    data = extract_features(file_name)
    features.append([file_name, data, instrument, pitch])
```

```python
features_df = pd.DataFrame(features, columns=["file", "feature", "instrument", "pitch"])
```

```python
features_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
X = np.array(features_df.feature.tolist())
y = np.array(features_df.pitch.tolist())

le = LabelEncoder()
yy = le.fit_transform(y)
```

```python
le.classes_
```

```python
len(le.classes_)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, yy, test_size=0.2, shuffle=True, random_state=42)
```

```python
len(x_test)
```

```python
len(y_test)
```

```python
from sklearn.svm import SVC
from sklearn import metrics
```

```python
model = SVC(kernel="linear")
model.fit(x_train, y_train)
```

```python
y_pred_train = model.predict(x_train)
print("Train Accuracy:", metrics.accuracy_score(y_train, y_pred_train))

y_pred_test = model.predict(x_test)
print("Test Accuracy:", metrics.accuracy_score(y_test, y_pred_test))
```

```python
model = Sequential()
model.add(Conv1D(32, 1, strides=1, activation='relu', input_shape=(40, 1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Conv1D(64, 1, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
# model.add(Dense(128, activation='relu'))
model.add(Flatten())
model.add(Dense(12, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
from datetime import datetime

num_epochs = 100
batch_size = 32

checkpointer = ModelCheckpoint(filepath="best_weights.hdf5", verbose=1, save_best_only=True)

start = datetime.now()
model.fit(Xtrain, y_train, batch_size=batch_size, epochs=num_epochs, validation_data=(x_test, y_test), 
          callbacks=[checkpointer], verbose=1)


duration = datetime.now() - start
print("Training completed in: ", duration)
```

```python
train_score = model.evaluate(x_train, y_train, verbose=1)
print("Training Accuracy: ", train_score[1])
```

```python
test_score = model.evaluate(x_test, y_test, verbose=1)
print("Test Accuracy: ", test_score[1])
```
